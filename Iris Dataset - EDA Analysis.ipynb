```python
iris_dist = iris_data.groupby('Species').size()
print("There are following observerations in each category\n",iris_dist)
```

```python
iris_dist.plot(title="Species Distribution")
```

```python
iris_corr = iris_data.corr()
print(iris_corr)
```

Exploratory Data Analysis using Visualization
------------------------------------------------

```python
#Scatter plot for Sepal Length Vs Width

figure = iris_data[iris_data.Species=='Iris-setosa'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',label='setosa')
iris_data[iris_data.Species=='Iris-versicolor'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',ax=figure,color='red',label='versicolor')
iris_data[iris_data.Species=='Iris-virginica'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',ax=figure,color='magenta',label='virginica')

figure.set_title("Sepal Length Vs Width")
```

```python
#Scatter plot for Petal Length Vs Width

figure = iris_data[iris_data.Species=='Iris-setosa'].plot(kind='scatter',x='PetalLengthCm',y='PetalWidthCm',label='setosa')
iris_data[iris_data.Species=='Iris-versicolor'].plot(kind='scatter',x='PetalLengthCm',y='PetalWidthCm',ax=figure,color='red',label='versicolor')
iris_data[iris_data.Species=='Iris-virginica'].plot(kind='scatter',x='PetalLengthCm',y='PetalWidthCm',ax=figure,color='magenta',label='virginica')

figure.set_title("Petal Length Vs Width")
```

Observation:

As we can see that the Petal Features are giving a better cluster division compared to the Sepal features. This is an indication that the Petals can help in better and accurate Predictions over the Sepal

```python
iris_data.hist(edgecolor='black')
```

```python

sns.FacetGrid(iris_data, hue="Species") \
   .map(sns.distplot, "PetalLengthCm") \
   .add_legend();
sns.FacetGrid(iris_data, hue="Species") \
   .map(sns.distplot, "PetalWidthCm") \
   .add_legend();
sns.FacetGrid(iris_data, hue="Species") \
   .map(sns.distplot, "SepalWidthCm") \
   .add_legend();
sns.FacetGrid(iris_data, hue="Species") \
   .map(sns.distplot, "SepalLengthCm") \
   .add_legend();
```

```python
sns.heatmap(iris_corr,annot=True)
```
